```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
a = np.arange(1, 101)
b = np.arange(501, 601)
```

```python
print(a)
print(b)
```

```python
train_test_split(a)
```

```python
a_train, a_test = train_test_split(a, test_size=0.2)
```

```python
a_train.shape, a_test.shape
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2)
```

```python
print(a_test)
print()
print(b_test)
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=40)
print(a_train)
print()
print(a_test)
print()
print(b_train)
print()
print(b_train)
print()
```
