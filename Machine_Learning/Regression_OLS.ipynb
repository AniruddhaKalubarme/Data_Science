```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'Marks':         [42, 48, 53, 61, 66, 72, 79, 83],
    'Random':        [1,3,2,1,3,2,1,3]
})

print(df)
```

```python
x = sm.add_constant(df['Hours_Studied'])
```

```python
result = sm.OLS(df['Marks'], x).fit()
```

```python
result.summary()
```

```python
y_predicted = 6*df['Hours_Studied'] + 36
```

```python
print(y_predicted)
```

```python
plt.scatter(df['Hours_Studied'], df['Marks'], label='Actual Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.plot(df['Hours_Studied'], y_predicted, c='green', label='Predicted Marks')
plt.legend()
```

```python
y = df['Marks']
x1 = df[['Hours_Studied', 'Random']]
x = sm.add_constant(x1)
```

```python
result = sm.OLS(y, x).fit()
```

```python
result.summary()
```
