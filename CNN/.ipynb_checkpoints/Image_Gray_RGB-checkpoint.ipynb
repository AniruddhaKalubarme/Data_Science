```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('shivaji_maharaj.jpg')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

```python
plt.imshow(R, cmap='Reds')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,5))
ax1.imshow(R, cmap='Reds')
ax2.imshow(G, cmap='Greens')
ax3.imshow(B, cmap='Blues')
plt.show()
```

```python
img
```

```python
img2 = img.copy()
```

```python
img2[:,:,2] = 255
```

```python
plt.imshow(img2)
```

```python
img2 = img.copy()
img2[100:500,100:500,2] = 255
plt.imshow(img2)
```

```python
img2 = img.copy()
img2[600:810,0:1120,:] = 0
img2[0:50,0:200,:] = 0
plt.imshow(img2)
```

# Convert an image to GrayScale

```python
imgGray = 0.333*R + 0.333*G + 0.333*B
print(imgGray)
plt.imshow(imgGray, cmap='gray')
```

```python
import cv2
imGrayCV2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imGrayCV2, cmap='gray')
```
