```python
num_cols = wdi.select_dtypes(include=['number']).columns
cat_cols = wdi.select_dtypes(exclude=['number']).columns
```

```python
wdi[num_cols] = wdi[num_cols].fillna(wdi[num_cols].mean())
```

```python
wdi.isnull().sum()
```

```python
wdi['alcohol_consumption_per_capita'].value_counts(dropna=False)
```

```python
wdi['country_category'].value_counts(dropna=False)
```

```python
from sklearn.impute import SimpleImputer
simple_imp_num=SimpleImputer(strategy='constant', fill_value=-999)
```

```python
imputed_num = simple_imp_num.fit_transform(wdi[num_cols])
```

```python
wdi[num_cols] = imputed_num
```

```python
simple_imp_cat=SimpleImputer(strategy='most_frequent')
```

```python
imputed_cat = simple_imp_cat.fit_transform(wdi[['country_category']])
```

```python
print(imputed_cat)
```

```python
wdi['country_category'] = imputed_cat.flatten()
```

```python
wdi.info()
```

```python
wdi.info()
```

```python
type(cat_cols)
```

```python
most_freq = wdi[cat_cols].describe().loc['top']
```

```python
wdi[cat_cols] = wdi[cat_cols].fillna(most_freq)
```

```python
wdi['country_category'].value_counts()
```

```python
wdi.info()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
iter_imp = IterativeImputer(min_value=wdi[num_cols].min(), max_value=wdi[num_cols].max())
```

```python
imputed_data = iter_imp.fit_transform(wdi[num_cols])
```

```python
wdi[num_cols] = imputed_data
```
