```python
df['TEMP_Category'] = pd.qcut(df['TEMP'], q=5, labels=['Very Cold', 'Cold', 'Warm', 'Hot', 'Very Hot'])
```

```python
df[['TEMP', 'TEMP_Category']]
```

```python
df.columns
```

```python
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
df['day_of_week'] = df['date_time'].dt.dayofweek
```

```python
df['day_of_week']
```

```python
df['day_of_week'] = df['day_of_week'].map({
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday"
})
```

```python
df['day_of_week'].value_counts()
```

```python
df['day_of_week'].map(len)
```

```python
df['Weekend_Weekday'] = df['day_of_week'].map(lambda day:'Weekend' if day in ['Saturday', 'Sunday'] else 'Weekday')
```

```python
df['Weekend_Weekday'].value_counts()
```

```python
def WeekFind(day):
    if (day in ['Saturday', 'Sunday']):
        return 'Weekend'
    else:
        return 'Weekday'

df['day_of_week'].map(WeekFind)
```

```python
df['wd'].value_counts()
```

```python
def wd_direction(wd):
    if wd[0] == 'N':
        return 'North'
    elif wd[0] == 'S':
        return 'South'
    elif wd[0] == 'E':
        return 'East'
    else:
        return 'West'
df['wd'].map(wd_direction)
```

```python
df['wd'].map(wd_direction).value_counts()
```

```python
df[['PM2.5', 'PM10']].apply('mean', axis=0)
```

```python
df[['PM2.5', 'PM10']].mean()
```

```python
df[['PM2.5', 'PM10']].mean(axis=1)
```

```python
df.columns
```

```python
def activity_decision(pm25_category, temp_category, rain):
    if pm25_category in ['Good', 'Moderate'] and temp_category in ['Warm', 'Hot'] and rain == 0:
        return 'Go outside'
    else:
        return 'Stay inside'

df['activity'] = df.apply(lambda row: activity_decision(row['PM2.5_category'], row['TEMP_Category' ], row['RAIN']), axis=1)
```

```python
df[['activity', 'PM2.5_category', 'TEMP_Category', 'RAIN']]
```
