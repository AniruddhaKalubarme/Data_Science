```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Jack'],
    'Age': [20, 21, 25]
})

df2 = pd.DataFrame({
    'Name': ['Charlie', 'David', 'Jack'],
    'Age': [22, 23, 25]
})

result = pd.concat([df1, df2], ignore_index=True)
print(result)
```

```python
df3 = pd.DataFrame({'Name':['Alice','Bob']})
df4 = pd.DataFrame({'Age':[20,21]})

result = pd.concat([df3, df4], axis=1)
print(result)
```

```python
df5 = pd.DataFrame({
    'ID':[1,2,3],
    'Name':['A','B','C']
})

df6 = pd.DataFrame({
    'ID':[2,3,4],
    'Marks':[80,90,70]
})

result = pd.merge(df5, df6, on='ID')
print(result)
```

```python
print(pd.merge(df1, df2, on='ID', how='inner'))
```

```python
print(pd.merge(df1, df2, on='ID', how='left'))
```

```python
print(pd.merge(df1, df2, on='ID', how='right'))
```

```python
print(pd.merge(df1, df2, on='ID', how='outer'))
```
