```python
import pandas as pd
```

```python
data = pd.read_csv("air_quality_updated.csv")
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
df.agg({
    'PM2.5': ['min', 'max', 'mean', 'nunique'],
    'PM2.5_category': ['min', 'max', 'nunique']
})
```

```python
def max_minus_min(s):
    return s.max()-s.min()

max_minus_min(df['TEMP'])
```

```python
df[['PM2.5', 'TEMP']].agg(['min', 'max', max_minus_min])
```

```python
df[['PM2.5', 'PM10']].min(axis=1)
```

```python
df[['PM2.5', 'PM10']].mean(axis=1)
```

```python
df[['PM2.5', 'PM10']].sum(axis=1)
```

```python
df.groupby(by='PM2.5_category').groups
```

```python
df.groupby(by='PM2.5_category').groups.keys()
```

```python
df.groupby(by='PM2.5_category').get_group('Good')
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df['year'] = df['date_time'].astype('datetime64[ns]').dt.year
df.sort_values('date_time').groupby('year').first()
```

```python
df.groupby('TEMP_Category').size()
```

```python
df['TEMP_Category'].value_counts(sort=False)
```

```python
df['quarter'] = df['date_time'].astype('datetime64[ns]').dt.quarter
```

```python
df[['PM2.5', 'TEMP', 'quarter']].groupby('quarter').mean()
```

```python
df.groupby('quarter')[['PM2.5', 'TEMP']].mean()
```

```python
df.groupby('quarter')[['PM2.5', 'TEMP']].describe()
```

```python
df.groupby('quarter')[['PM2.5', 'TEMP']].agg(['min', 'max'])
```

```python
df.groupby('day_of_week')[['PM2.5', 'TEMP', 'RAIN']].agg({
    'PM2.5':['min', 'max', 'mean'],
    'TEMP':'mean',
    'RAIN':'mean'
})
```

```python
df.groupby(['quarter', 'TEMP_Category'])[['PM2.5', 'TEMP']].mean()
```
