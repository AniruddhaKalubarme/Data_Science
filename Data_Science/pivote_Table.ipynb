```python
import pandas as pd
```

```python
data = pd.read_csv('student.csv')
student = pd.DataFrame(data)
```

```python
student.columns
```

```python
student.head()
```

```python
pd.pivot_table(student,
              index='sex',
              values='score')
```

```python
pd.pivot_table(student,
              index=['sex', 'internet'],
              values=['score', 'age'],
              aggfunc=['min', 'mean', 'max'])
```

```python
pd.pivot_table(student,
              index=['sex', 'internet'],
              values='score',
              aggfunc='mean',
              columns='studytime')
```

```python
pd.pivot_table(student,
              index=['sex', 'internet'],
              values=['score', 'age'],
              aggfunc='mean',
              columns='studytime',
              fill_value=-999)
```

```python
pd.pivot_table(student,
              index=['sex', 'internet'],
              values=['score'],
              aggfunc='mean',
              columns=['studytime', 'internet'],
              fill_value=-999)
```

```python
pd.pivot_table(student,
               index='familysize',
               values='score',
               columns='sex')
```

```python
pd.pivot_table(student,
               index='familysize',
               values='score',
               columns='sex',
              margins=True,
              margins_name='Average sscore total')
```

```python
pd.pivot_table(student,
               index='studytime',
               columns='sex',
               aggfunc={'age':['min', 'max'],
                        'score': 'median'})
```

```python
pd.pivot_table(student,
               index='studytime',
               values='score',
               columns='sex',
               aggfunc=lambda s:s.max()-s.min())
```
