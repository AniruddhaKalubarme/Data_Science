```python
Q1 = wdi['population'].quantile(0.25)
Q2 = wdi['population'].quantile(0.50)
Q3 = wdi['population'].quantile(0.75)

print(Q1, Q2, Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
U_limit = 1.5*IQR+Q3
L_limit = Q1 - 1.5*IQR
```

```python
outliers = wdi[(U_limit<wdi['population']) | (L_limit>wdi['population'])]
```
